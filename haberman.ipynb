```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
col_names = ["age","operation_year","axillary_nodes","survival_status"]
data = pd.read_csv("/Users/debdeepsanyal/Downloads/haberman.csv", names = col_names)
```

```python
print(data.shape)
print(data.columns)
```

```python
print(data.head())
```

```python
print("Here's some high level statistics for the dataset \n")
print("Number of data points : ", len(data))
print("Number of features : ", data.shape[1] - 1) #excluding the label
print("Number of classes : ", data["survival_status"].nunique()) 
print("Data points per class : ")
print(data['survival_status'].value_counts())
```

```python
print("The objective of the EDA is to understand the characteristics of the data and identify features useful for classification")
```

```python
print("Univariate Analysis ")
sns.FacetGrid(data, hue = "survival_status", height = 5).map(sns.histplot, "age").add_legend()
plt.title("PDF for age ")
plt.xlabel("age")
plt.ylabel("Count")
plt.show()
```

```python
#For operation year 
sns.FacetGrid(data, hue = 'survival_status', height = 5).map(sns.histplot, "operation_year").add_legend()
plt.title("PDF for operation year ")
plt.xlabel("Operation year")
plt.ylabel("Count")
plt.show()
```

```python
#PDF and CDF for Axillary Nodes
sns.FacetGrid(data, hue = "survival_status", height = 5).map(sns.histplot, "axillary_nodes").add_legend()
plt.title("PDF for Axillary Nodes")
plt.xlabel("Axillary nodes")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x = "survival_status", y = "age", data = data)
plt.title("Box plot for age ")
plt.xlabel("Survival Status")
plt.ylabel("Age")
plt.show()
```

```python
sns.boxplot(x = "survival_status", y = "axillary_nodes", data = data)
plt.title("Box plot for axillary nodes")
plt.xlabel("Survival Status")
plt.ylabel("Axillary Nodes")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.pairplot(data, hue = "survival_status", height = 5)
plt.title("Pair plot")
plt.show()
```
